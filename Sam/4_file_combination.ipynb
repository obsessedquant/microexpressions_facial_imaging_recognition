# File combination

This code is meant to join the augmented datasets together.

```python
import pandas as pd

df0 = pd.read_csv('./Data/0_angry_augmented.csv')
df0.drop('Unnamed: 0',axis=1,inplace=True)

df1 = pd.read_csv('./Data/1_disgust_augmented.csv')
df1.drop('Unnamed: 0',axis=1,inplace=True)

df2 = pd.read_csv('./Data/2_fear_augmented.csv')
df2.drop('Unnamed: 0',axis=1,inplace=True)

df5 = pd.read_csv('./Data/5_surprise_augmented.csv')
df5.drop('Unnamed: 0',axis=1,inplace=True)

result = pd.concat([df0, df1, df2, df5])
result
```

```python
result.to_csv('./Data/augmented_data_compiled.csv')
```
